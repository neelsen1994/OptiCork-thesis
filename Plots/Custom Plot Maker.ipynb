```python
import numpy as np
import matplotlib.pyplot as plt


data = np.genfromtxt('./Plots/Training_All_NoAug_100epo/epoch_loss.csv', delimiter = ',')
data2 = np.genfromtxt('./Plots/Training_All_NoAug_100epo/epoch_val_loss.csv', delimiter = ',')
print(data[1:,2])

plt.figure(figsize=(30, 8))

# Plot training & validation loss values
plt.subplot(122)
plt.plot(data[1:,1],data[1:,2])
plt.plot(data2[1:,1],data2[1:,2])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train','Validation'], loc='upper left')
plt.grid(True)
plt.show()
```
